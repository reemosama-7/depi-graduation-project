```python
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
```

```python
print(df['birth_date'])
```

```python
def calculate_age(birth_date):
    if pd.isnull(birth_date):  
        return None
    today = date.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return int(age)

df['age'] = df['birth_date'].apply(calculate_age)

print(df[['birth_date', 'age']])
```

```python
df
```

```python
df['age'] = df['age'].fillna(0)
df['age'] = df['age'].astype(int)
print(df[['birth_date', 'age']])
```

```python
plt.hist(df.medal_type, color='#ebc634')
plt.ylabel('Number of medals')
plt.title('Count of Medals')
plt.show()
```

```python
top_5_disciplines = df['discipline'].value_counts().nlargest(5).index
df_top_5 = df[df['discipline'].isin(top_5_disciplines)]
plt.hist(df_top_5['discipline'], color='#498028')
plt.ylabel('Number of medals')
plt.title('Count of Medals (Top 5 Disciplines)')
plt.xticks(rotation=45)  # Optional: rotate labels for better readability
plt.show()
```

```python
df_top_5_oldest = df.nlargest(5, 'age')

plt.hist(df_top_5_oldest['age'], color='#c40e14')
plt.ylabel('Count of players')
plt.title('Oldest Participants')
plt.xticks(rotation=45)
plt.show()
```

```python
female_count = df.loc[df['gender'] == 'Female'].shape[0]
male_count = df.loc[df['gender'] == 'Male'].shape[0]
sizes = [female_count, male_count]
labels = ['Female', 'Male']
colors = ['#ff9999', '#2995e3']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Gender Distribution in Olympics')
plt.show()
```

```python
grouped_data = df.groupby(['discipline', 'gender']).size().unstack(fill_value=0)
grouped_data['Total'] = grouped_data.sum(axis=1)

top_5_disciplines = grouped_data.nlargest(5, 'Total').drop(columns='Total')
top_5_disciplines.plot(kind='bar', stacked=True, color=['#ff9999', '#2995e3'])
plt.title('Gender Distribution by Discipline (Top 5)')
plt.xlabel('Discipline')
plt.ylabel('Number of Participants')
plt.legend(['Female', 'Male'], loc='upper right')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
